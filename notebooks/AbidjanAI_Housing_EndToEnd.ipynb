7. Model Evaluation & Comparison

Random Forest achieved the best generalization (lowest RMSE, highest R²).

Decision Tree showed signs of overfitting (worse test metrics than RF).

8. Hyperparameter Tuning

RandomizedSearchCV was applied to optimize the Random Forest model.

However, tuning did not lead to a performance improvement:

Baseline RMSE: 0.4997

Tuned RMSE: 0.5006

Baseline R²: 0.8094

Tuned R²: 0.8088

The tuned model performed slightly worse than the baseline model. This suggests that:

The default Random Forest parameters were already well-suited to the dataset.

The search space may require refinement.

The model may already be near its performance ceiling given the dataset characteristics (including the capped target at 5.0).

Therefore, the baseline Random Forest model was selected as the final model.

9. Final Result & Conclusion

This project implemented a complete end-to-end machine learning pipeline for housing price prediction.

Key findings:

Median income (MedInc) is the strongest predictor of housing value.

Geographic features significantly influence pricing due to spatial clustering.

Among all tested models, Random Forest achieved the best generalization performance, with:

RMSE = 0.4997

R² = 0.8094

Hyperparameter tuning using RandomizedSearchCV did not improve performance beyond the baseline model, indicating that the default Random Forest configuration was already well-suited for this dataset.

The final selected model is the Baseline Random Forest Regressor.

Limitations

The target variable is artificially capped at 5.0, which may limit prediction accuracy for high-value houses.
